#### Skill Challenge

###### **1.**

Starting with our main dataframe (*sat*), create a pivot table that summarizes the <ins>total</ins> student enrollment (Student Enrollment) across all 5 boroughs (*Borough*). 

Which boroughs have the highest and lowest high school student bodies?

###### **2.**

Modify the pivot table from the step above to also reflect the average student body size (*Student Enrollment*) across boroughs, in the same pivot table.

###### 🔥🔥 🚒 **3.**

Create a pivot table of high schools from the Queens borough (*Borough*), containing the SAT section scores (*SAT Section*, *Score*) as columns and the school name (*School Name*) as index. Sort the table in descending order by math section scores.

#### Solution

```
sat.head()
```

```
# 1
```

```
sat.pivot_table('Student Enrollment', 'Borough', aggfunc='sum').sort_values(by='Student Enrollment', ascending=False)
```

```
sat.pivot_table('Student Enrollment', 'Borough', aggfunc='sum').nlargest(50, columns='Student Enrollment')
```

```
sat.pivot_table('Student Enrollment', 'Borough', aggfunc='sum').idxmin()
```

```
sat.pivot_table('Student Enrollment', 'Borough', aggfunc='sum').idxmax()
```

```
# 2
```

```
sat.pivot_table('Student Enrollment', 'Borough', aggfunc=['sum', 'mean'])
```

```
# 3
```

```
# - queens schools
# - a pivot table of scores (by sat section), school names
# - sorted descending by math section scores
```

```
sat[sat.Borough == 'Queens']\
  .pivot_table('Score', 'School Name', 'SAT Section')\
  .sort_values(by='Math', ascending=False)
```
