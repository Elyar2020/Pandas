# 92:Filtering IN 2D

```python
nutrition.shape
```

```python
nutrition.filter(like = "octopus",axis =0) # like parameter is case sensitive. with capital Octopus we get diffrent result.
```

```python
# Filter with regex:
nutrition.filter(regex= "octopus", axis = 0) # regex parameter is casesensitive as well 
```

```python
# This returns any row with octupus either it start with upper case or lower case o.
nutrition.filter(regex= "[Oo]ctopus",axis =0) 
```

```python
# 2ND option: in regex (?i) means not case sensitive 
nutrition.filter(regex ="(?i)octopus",axis = 0)
```

# Filttering along both Dime

```python
# First option:
nutrition.filter(regex = "(?i)octopus",axis = 0)\
    .filter(items = ["cholesterol_mg","serving_size_g","calories"],axis = 1)
```

```python
#2ND: option:
nutrition.filter(regex= "(?i)octopus",axis = 0)\
          .loc[:, ["cholesterol_mg","serving_size_g","calories"]]
```
